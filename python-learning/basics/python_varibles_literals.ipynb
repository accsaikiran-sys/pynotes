```python
# Variable assignment
count = 10
price = 99.99
name = "Python"

print(count)   # Output: 10
print(price)   # Output: 99.99
print(name)    # Output: Python 
```

```python
a, b,c  = 1 ,2, 3
print(a)
print(b)
print(c)
```

```python
value = 100
print(value)
value = "none"
print(value) #reassigned
```

```python
integer_literal = 42
float_literal = 3.14
complex_literal = 2 + 3j

print(integer_literal)  # Output: 42
print(float_literal)    # Output: 3.14
print(complex_literal)  # Output: (2+3j)
```

```python
single_quote = 'Hello'
double_quote = "World"
multi_line = """This is
a multi-line
string"""

print(single_quote)  
print(double_quote)  
print(multi_line)
```

```python
is_active = True
is_logged_in = False

print(is_active)     
print(is_logged_in)   
```
