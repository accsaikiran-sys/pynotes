```python
fruits = ["apple", "banana", "orange"]

for fruit in fruits:
    print(fruit)
```

```python
for i in range(5):
    print(i)
```

```python
for i in range(2, 10, 2):
    print(i)
```

```python
word = "Python"

for char in word:
    print(char)
```

```python
languages = ["Python", "Java", "C++"]

for index, lang in enumerate(languages):
    print(index, lang)
```

```python
for i in range(3):
    for j in range(2):
        print(f"i={i}, j={j}")
```

```python
for number in range(1, 10):
    if number == 5:
        break
    print(number)
```

```python
for number in range(1, 6):
    if number == 3:
        continue
    print(number)
```

```python
for num in range(10):
    if num == 3:
        continue
    # if num == 5:
    #     break
    print(num)
else:
    print("Loop completed successfully")
```

```python
student = {"name": "Alice", "age": 22, "grade": "A"}

for key, value in student.items():
    print(key, ":", value)
```
